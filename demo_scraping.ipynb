```python
def make_it_lower(array):
    
    new_array = []
    for item in array:
        new_array.append(item.lower())
        
    return new_array
```

```python
#now lets get some more info: the column names
def bs_column_names(tag):
    return tag.name == 'th' and tag.has_attr('width') #custom BeautifulSoup search

def get_column_names(namedata):
    column_names = []
    for name in namedata.find_all(bs_column_names):
        try:
            match = re.search('([A-Z3].*[A-Z])', str(name)) #refine the matches by regex
            #if name.get_text() not in column_names: 
            if match.group(0).strip() not in column_names: 
                column_names.append(match.group(0).strip())
        except:
            a = 1
            
    return column_names
```

```python
data_columns = get_column_names(bit1)
print data_columns
data_columns = make_it_lower(data_columns)   
print data_columns
```

```python
def bs_column_data(tag):
    return tag.name == 'td' and not tag.has_attr('style') #another custom BeautifulSoup search

def get_column_data(datadata):
    column_data = []
    for data in datadata.find_all(bs_column_data):
        try:
            match = re.search('[0-9]*[0-9\-]*[0-9\-]*[0-9\-]', str(data)) #matching a number pattern
            column_data.append(match.group(0).strip()) #removing whitespace
        except:
            a = 1 
            
    return column_data
```

```python
column_data = get_column_data(bit1)
print column_data
```
